## lmfit

```python
import lmfit #lmfit套件
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
```

```python
# 產生資料

x = np.linspace(0,10,100)
y = np.sin(5*x)*np.exp(-x/2.5)
y += npr.choice([-1,1], size=y.shape)*npr.random(size=y.shape)/5
```

```python
# 資料繪圖

%matplotlib notebook

plt.figure()
plt.plot(x,y,'*')
plt.show()
```

## model fitting

```python
# mdoel

model = lmfit.models.ExpressionModel("ampl * sin((x - x0)*freq) * exp(-x/tau) + offset") # create model
params = model.make_params(ampl=1, x0=0, freq=10, tau=1, offset=0) #give guess parameters
params["offset"].set(vary=False)
```

```python
fit = model.fit(y, params, x=x) # input x,y & guess parameters
#fit = model.fit(y, params, x=x, weights=) # input x,y,guess parameters & weight(error)
```

```python
# output fit's report

print(fit.fit_report())
#print(fit.covar)
```

```python
plt.figure()
plt.plot(x,y,'*')
plt.plot(x, fit.best_fit,'r-')
plt.show()
```
